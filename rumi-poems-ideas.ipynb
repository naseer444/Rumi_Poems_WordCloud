```python
import pandas as pd
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('../input/rumi-poems/rumi_poetry.csv')

# with open('../input/rumi-poems/rumi_poetry.csv', 'rU') as myfile:
#     filtered = (line.replace('\n', '') for line in myfile)
#     for row in csv.reader(filtered)
comment_words = ''
stopwords = set(STOPWORDS)
```

```python
data.head(5)
```

## Some Preprocessing

```python
for val in data.Poem:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
```

```python
wordcloud = WordCloud(width = 1000, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
```

## Let's draw the computed WordCloud

```python
# plot the WordCloud image                      
plt.figure(figsize = (11, 11), facecolor = None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("on")
plt.tight_layout(pad = 0)
 
plt.show()
```

## Let's Store the image as png file

```python
wordcloud.to_file('N.png')
```
